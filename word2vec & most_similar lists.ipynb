```python
import sys
import gensim, logging

logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
#данные на вход
f = 'corpus.txt'
data = gensim.models.word2vec.LineSentence(f)
```

```python
#инициализируем
model = gensim.models.Word2Vec(data, size=500, window=10, min_count=2, sg=0)
```

```python
#нормализуем
model.init_sims(replace=True)
#print(len(model.vocab))
model.save('my.model')
```

```python
m = 'my.model'
if m.endswith('.vec.gz'):
    model = gensim.models.Word2Vec.load_word2vec_format(m, binary=False)
elif m.endswith('.bin'):
    model = gensim.models.KeyedVectors.load_word2vec_format(m, binary=True)
else:
    model = gensim.models.Word2Vec.load(m)
```

```python
words_food = ['еда', 'рыба']
words_service = ['официант', 'персонал']
words_interior = ['обстановка', 'атмосфера']
```

```python
list_food = []
list_service = []
list_interior = []

def list_of_similar(words, list_to_save_at):
    for word in words:
        # есть ли слово в модели? Может быть, и нет
        if word in model:
            #print(word)
            # смотрим на вектор слова (его размерность 300, смотрим на первые 10 чисел)
            #print(model[word][:10])
            # выдаем 10 ближайших соседей слова:
            for i in model.most_similar(positive=[word], topn=40):
                # слово 
                list_to_save_at.append(i[0])
    return set(list_to_save_at)
            
list_food  = list_of_similar(words_food, list_food)
list_service = list_of_similar(words_service, list_service)
list_interior = list_of_similar(words_interior, list_interior)
```

```python
from collections import Counter
import json

with open('seed.txt', 'r', encoding='utf-8') as f:
    f = f.readlines()
    words = [i.strip() for i in f]
allwords = []
for word in words:
    # есть ли слово в модели? Может быть, и нет
    if word in model:
        print(word)
        # смотрим на вектор слова (его размерность 300, смотрим на первые 10 чисел)
        #print(model[word][:10])
        # выдаем 10 ближайших соседей слова:
        for i in model.most_similar(positive=[word], topn=50):
            # слово + коэффициент косинусной близости
            allwords.append(i[0])
    else:
        # Увы!
        print(word + ' is not present in the model')
      
words_c = Counter(allwords)
with open('file.txt', 'w', encoding='UTF-8') as file:
    file.write(json.dumps(words_c))
```
