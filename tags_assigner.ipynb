```python
import json

with open("cat_polarized.json", "r") as io:
    words = json.load(io)
```

```python
#[0] - mean, [1] - отклонение, 
#среднее > 7, отклонение <2 - pos
#среднее < 7, отклонение <2 - neg
#отклонение > 3 - ambi

to_results = {}

def accuracy(words):
    acc = 0
    if words[0] >= 7 and words[1] < 2:
        acc = 0.8
        
    if words[0] >= 7 and words[1] < 2.5 and words[1] > 2:
        acc = 0.5
        
    elif words[0] < 5 and words[1] < 2.5:    
        acc = 0.8
    
    elif words[0] > 5 and words[0] >= 7 and  words[1] < 2.5 and words[1] > 2:    
        acc = 0.5

    if words[1] > 3 or words[2] < 25:  
        acc = 0.5

    if words[1] >= 2.5 and words[1] <= 3:
        acc = 0.3
    
    return acc

def tagging(words):
    #tag = 'non-tematic'
    if words[0] >= 7 and words[1] < 2.5:
        tag = 'pos'
        
    elif words[0] < 7 and words[1] < 2.5:    
        tag = 'neg'

    elif words[1] > 2.5:  
        tag = 'ambi'

    #elif words[1] >= 2.5 and words[1] <= 3:
    #    tag = 'non-tematic'      
    return tag



for param in words.keys():
    for word in words[param].keys():
        
        if words[param][word][2] > 10:            
        
            tag = tagging(words[param][word])
            acc = accuracy(words[param][word])
            
            if word in to_results:
                if to_results[word][3] == tag:
                    to_results[word][4] += (' & ' + param)
                elif to_results[word][3] != tag and tag != 'ambi':
                    to_results[word+'_'+param] = [str(words[param][word][0]), str(words[param][word][1]), str(words[param][word][2]), tag, param, str(acc)]
            
            else:
                to_results[word] = [str(words[param][word][0]), str(words[param][word][1]), str(words[param][word][2]), tag, param, str(acc)]
            
            #elif tag == 'non-tematic' and word not in to_results:
            #to_results[word] = [str(words[param][word][0]), str(words[param][word][1]), str(words[param][word][2]), tag, '']
sorted_results = OrderedDict(sorted(to_results.items(), key=lambda x: x[1][-1], reverse=True))
```

```python
with open("results.csv", "w") as io:
    io.write("word,mean,sd,freq,tag,area,acc\n")
    
    for word in sorted_results.keys():
        io.write(word + ',' + ','.join(sorted_results[word]) + "\n")
```
